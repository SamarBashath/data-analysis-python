```python
import re
def extract_email_ID(string):
    email = re.findall(r'<(.+?)>', string)
    if not email:
        email = list(filter(lambda y: '@' in y, string.split()))
    return email[0] if email else np.nan
```

```python
mail['from'] = mail['from'].apply(lambda x: extract_email_ID(x))
```
